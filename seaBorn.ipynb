```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
print(df.head())
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.lineplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
sns.countplot(x='day', data=df)
plt.show()

```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.show()
```

```python
sns.set_style('darkgrid')  # white, whitegrid, dark, darkgrid, ticks
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.show()
```

```python
g = sns.FacetGrid(df, row='sex', col='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
sns.set_palette("coolwarm")
sns.boxplot(x='day', y='tip', data=df)
plt.show()
```

```python
sns.set_style('darkgrid')  # options: white, dark, darkgrid, whitegrid, ticks
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df, color='red')
plt.show()
```

```python
sns.scatterplot(x='day', y='total_bill', data=df, color='#33cc33')
```

```python
plt.figure(figsize=(4, 1))
sns.barplot(x='day', y='total_bill', hue='sex', data=df,
            palette={'Male': 'pink', 'Female': 'blue'})

plt.show()
```

```python
plt.title("Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip Amount")


sns.lineplot(x='total_bill', y='tip', data=df, marker='D',color="red",errorbar=None,linestyle='--',linewidth='3')  # ci percentage can be controlled too
```

```python

plt.gca().set_facecolor('#f0f0f0')
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)

plt.legend(loc='upper right')  # options: best, lower right, etc.
plt.show()
```
