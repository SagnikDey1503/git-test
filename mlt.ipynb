```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.savefig("plo1t.png")
```

```python
x = [1, 2, 3]
y1 = [1, 4, 9]
y2 = [1, 2, 3]

plt.subplot(1,2, 1)
plt.plot(x, y1,label="square")
plt.title("Square")
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, y2,label="linear")
plt.title("Linear")

plt.tight_layout()
plt.legend()
plt.show()

```

```python
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins=, color='green', alpha=0.7)
plt.title("Histogram")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot")
plt.show()
```

```python
x=[1,2,3,4]
y=[4,6,7,9]
plt.figure(figsize=(4, 4), dpi=100)
plt.plot(x, y)
plt.xlim(0, 5)
plt.ylim(0, 10)
plt.title("Custom ")
plt.text(2, 4, "Point", fontsize=12, color='blue')
plt.show()
```
